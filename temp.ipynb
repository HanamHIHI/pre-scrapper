```python
cate_dict = {}

for i in range(len(df2)):
    cate_dict[df2.iloc[i]["name"]] = df2.iloc[i]["category"]

for key, val in cate_dict.items():
    print(key, val)
```

```python
df_pre['category2'] = df_pre["category"].copy().apply(lambda x: x.replace(x, x.replace(',', ' ').replace('"', '')))
df_pre
```

```python
df_pre["category3"] = df_pre['category2'].apply(lambda x : x.replace(x, cate_dict[x]))
# df_pre

df_post = df_pre[["name", "category3"]]
df_post.to_csv("category.csv")
```
